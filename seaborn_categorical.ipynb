```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# Categorical Scatterplots

- x-axis --> categorical data
- y-axis --> numerical data


```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

```python
# strip plot
sns.catplot(x='day',y='tip',kind='strip',data=tips)
```

```python
# swarm plot also bescribes the distribution of the data
sns.catplot(x="day",y="tip",kind="swarm",data=tips)
```

```python
sns.catplot(x="day",y="tip",kind="swarm",data=tips,hue="sex")
```

```python
sns.swarmplot(x='day',y='tip',data=tips)
```

## Box Plots

```python
sns.boxplot(tips['tip'])
```

```python
sns.catplot(x='day',y='total_bill',kind='box',data=tips)
```

```python
sns.catplot(x='day',y='total_bill',hue="sex",kind="box",data=tips)
```
