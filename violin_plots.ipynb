```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
tips=sns.load_dataset('tips')
```

violin plots are often used to compare the distribution of a given varialble across some categories

```python
sns.violinplot(tips['total_bill'])
```

```python
sns.catplot(x='day',y='total_bill',kind='violin',data=tips)
```

```python
sns.catplot(x='day',y='total_bill',kind='violin',data=tips,hue='sex')
```

```python
sns.catplot(x='day',y='total_bill',kind='violin',data=tips,hue='sex',split=True)
```
