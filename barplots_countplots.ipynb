```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
tips=sns.load_dataset('tips')
```

```python
sns.catplot(x='smoker',y='total_bill',kind="bar",data=tips)
```

```python
sns.catplot(x='smoker',y='total_bill',kind="bar",data=tips,hue='sex')
```

```python
sns.catplot(x='smoker',y='total_bill',kind="bar",data=tips,hue='sex',estimator=np.mean)
```

```python
sns.catplot(x='sex',kind='count',data=tips,hue='smoker')
```
